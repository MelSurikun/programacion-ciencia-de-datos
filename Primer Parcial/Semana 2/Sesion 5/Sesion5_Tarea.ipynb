## 1. Diferncias entre Jupyter Notebok y Google Colab

### - Jupyter Notebook
- **¿Qué es?:** Una aplicación de código abierto que corre localmente en tu computadora. Permite crear y ejecutar cuadernos interactivos con código (Python principalmente), texto con Markdown, gráficas, etc.

- **¿Dónde se ejecuta?:** En tu propia máquina (puedes instalarlo con Anaconda, pip o como extensión en VS Code).

- **Ventajas:**
   - Total control del entorno, librerias y configuración.
   - Puedes trabajar sin internet.
   - Ideal si necesitas usar datos grandes almacenados localmente.

- **Desventajas**
   - Necesitas instalar Python, librerias y configurar todo.
   - El rendimiento depende del hadwere de tu computadora.

### - Google Colab
- **¿Qué es?:** Una plataforma de Google basada en la nube que permite usar cuadernos muy similares a Jupyter (de hecho está basado en él).

- **¿Dónde se ejecuta?:** En los servidores de Google, a través de tu navegador.

- **Ventajas:**
   - No requiere instalación, solo tu cuenta de google.
   - Ofrece GPU y TPU gratis (aunque limitadas en tiempo).
   - Facil de compartir, como Google Docs.
   - Todo se guarda en Google Drive.

- **Desventajas**
   - Necesitas internet para usarlo.
   - Limitaciones de tiempo de ejecución (sesiones de 12 h aprox).
   - Menos control sobre la configuracion del sistema.

### - Nota
- Si trabajas con proyectos locales y pesados, Jupyter Notebook es mejor.
- Si quieres facilidad, colaboración y GPU gratis, Google Colab es más conveniente.

## 2. Ejemplo Practico con Pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv("airlines_flights_data.csv")

# Ver las primeras filas
print(df.head())

# Ver información general
print(df.info())

# Ver estadísticas básicas
print(df.describe())

# Gráfico: cantidad de vuelos por aerolínea
sns.countplot(data=df, x='airline')
plt.xticks(rotation=45)
plt.title("Cantidad de vuelos por aerolínea")
plt.show()
```

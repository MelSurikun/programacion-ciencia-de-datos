##### **üîπ Problema 1: Propiedades de un numero**
**Descripci√≥n:** Analizar s√≠ un numero es positivo o negativo, entero o decimal y par o impar.

```python
print("     -> Analizando n√∫meros")
print("Te dir√© s√≠ es positivo o negativo, entero o decimal, par o impar")
print("--------------------------------------------------------------------")

entrada = input("\nIngresa un numero:")

if entrada == "":
    print("\nNo escribiste ning√∫n numero. Usar√© el 5 como ejemplo")
    numero = 5.0
else:
    numero = float(entrada)

print("\nAnalizando...\n")

#positivo o negativo
if numero > 0:
    print("- Tu numero es POSITIVO")
elif numero < 0:
    print("- Tu numero es NEGATIVO")
else:
    print("- Tu numero es CERO")

#entero o decimal
if numero == int(numero):
    print("- Tu numero es ENTERO (no tiene decimales)")
    
    #par o impar solo a enteros
    numero_entero = int(numero)
    if numero_entero % 2 == 0:
        print("- Tu numero es PAR")
    else:
        print("- Tu numero es IMPAR")
        
else:
    print("- Tu numero es DECIMAL (tiene punto)")
    print("- N√∫meros decimales no son ni pares ni impares")

print("\nTU N√öMERO ES: ", entrada)

print("\n--------------------------------------------------------------------")
print("Fin del an√°lisis")
```

#### **üîπ Problema 3: Calcular IVA de una compra**
**Descripci√≥n:** Calculadora de cu√°nto se pagar√° de IVA en una compra, siendo alimentos b√°sicos del 0%, productos generales del 16% y productos de lujo del 25%.

```python
print("     -> Calculando IVA de compra")
print("Calcular IVA de compra de alimentos b√°sicos, productos generales o")
print("de productos de lujo.")
print("--------------------------------------------------------------------")

entrada_monto = input("¬øCu√°nto cuesta tu compra? $")
monto = float(entrada_monto)

print("\n¬øQu√© tipo de producto es?")
print("1. Alimentos b√°sicos (0% IVA)")
print("2. Productos generales (16% IVA)") 
print("3. Productos de lujo (25% IVA)")
entrada_tipo = input("Escribe el n√∫mero (1, 2 o 3): ")

if entrada_tipo == "1":
    iva = 0.00  # 0%
    tipo_producto = "Alimentos b√°sicos"
    print("\nUsaremos 0% de IVA (alimentos b√°sicos)")
elif entrada_tipo == "2":
    iva = 0.16  # 16%
    tipo_producto = "Productos generales"
    print("\nUsaremos 16% de IVA (productos generales)")
elif entrada_tipo == "3":
    iva = 0.25  # 25%
    tipo_producto = "Productos de lujo"
    print("\nUsaremos 25% de IVA (productos de lujo)")
else:
    iva = 0.16  # 16% por defecto
    tipo_producto = "Productos generales"
    print("\nN√∫mero no v√°lido. Usar√© 16% de IVA")

if monto <= 0:
    print("El monto debe ser mayor a cero. Usar√© $100")
    monto = 100

monto_iva = monto * iva
total_pagar = monto + monto_iva

print("\nCalculando...")
print("\n--------------------------------------------------------------------")
print("RESULTADO DE TU COMPRA")
print("--------------------------------------------------------------------")
print(f"- Tipo de producto: {tipo_producto}")
print(f"- Precio sin IVA: ${monto:.2f}")
print(f"- IVA aplicado: {iva*100}%")
print(f"- Monto de IVA: ${monto_iva:.2f}")
print(f"- - ->Total a pagar: ${total_pagar:.2f}")

print("\n--------------------------------------------------------------------")
print("Fin del c√°lculo")
```

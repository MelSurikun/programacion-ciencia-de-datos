##### **üîπ Ejercicio 1:** Adivinar el numero

```python
import random

print("     -> ADIVINANDO EL NUMERO")
print("--------------------------------------------------------------------")
print("pensar√© un n√∫mero del 1 al 50 y t√∫ lo adivinar√°s")

numero_secreto = random.randint(1, 50)
intentos = 0

while True:
    try:
        intento = int(input("\n¬øQu√© n√∫mero crees que es? (1-50): "))
        intentos += 1

        if intento < 1 or intento > 50:
            print("El n√∫mero debe estar entre 1 y 50, intenta de nuevo")
            
            continue

        if intento < numero_secreto:
            print("El numero que ingresaste es menor al que pens√©")
        elif intento > numero_secreto:
            print("El numero que ingresaste es mayor al que pens√©")
        else:
            print(f"\nAdivinasteeeeeeeeeeeeeeee en {intentos} intentos")
            print(f"El n√∫mero era: {numero_secreto}")
            break
            
    except ValueError:
        print("Ingresa solo n√∫meros")

print("\n--------------------------------------------------------------------")
print("Fin del juego")
```

##### **üîπ Ejercicio 2:** Lista de compras

```python
print("     -> LISTA DE COMPRAS")
print("--------------------------------------------------------------------")
print("Agrega productos a la lista de compras. Escribe 'fin' para terminar\n")

lista_compras = []
total_productos = 0

while True:
    producto = input("¬øQu√© producto quieres agregar? (escribe 'fin' para terminar): \n")
    
    if producto.lower() == 'fin':
        break
    
    if producto.strip() == "":
        print("Debes escribir un producto")
        continue
    
    lista_compras.append(producto)
    total_productos += 1
    print(f"Agregu√©: {producto}")

print("\n--------------------------------------------------------------------")
print("Lista de compras:\n")

if total_productos > 0:
    for i, producto in enumerate(lista_compras, 1):
        print(f"{i}. {producto}")
    
    print(f"\nTotal de productos: {total_productos}")
else:
    print("\nTu lista est√° vac√≠a")

print("\n--------------------------------------------------------------------")
print("Fin de la lista")
```

##### **üîπ Ejercicio 3:** Tablas de multiplicar

```python
print("     -> TABLAS DE MULTIPLICAR")
print("--------------------------------------------------------------------")
print("Te mostrar√© las tablas de multiplicar del 1 al 10:\n")

tabla_especifica = input("Quieres ver una tabla espec√≠fica? (si/no): ")

if tabla_especifica.lower() in ['s√≠', 'si', 's', 'yes']:
    try:
        numero = int(input("Qu√© tabla quieres ver? (1-10): "))
        if 1 <= numero <= 10:
            print(f"\nTABLA DEL {numero}:")
            print("--------------------------------------------------------------------")
            for i in range(1, 11):
                resultado = numero * i
                print(f"{numero} √ó {i} = {resultado}")
        else:
            print("El n√∫mero debe estar entre 1 y 10")
    except ValueError:
        print("Debes ingresar un n√∫mero")
else:
    print("\nTABLAS DEL 1 AL 10:")
    print("--------------------------------------------------------------------")
    
    for tabla in range(1, 11):
        print(f"\nTabla del {tabla}:")
        print("--------------------------------------------------------------------")
        
        for multiplicador in range(1, 11):
            resultado = tabla * multiplicador
            print(f"{tabla} √ó {multiplicador} = {resultado}")

print("\n--------------------------------------------------------------------")
print("Fin de las tablas")
```

##### **üîπ Ejercicio 4:** Calculadora de estad√≠sticas

```python
print("     ->CALCULADORA DE ESTAD√çSTICAS")
print("--------------------------------------------------------------------")
print("Ingresa n√∫meros y te dar√© estad√≠sticas, escribe 'fin' para terminar\n")

numeros = []
contador = 0

while True:
    entrada = input("Ingresa un n√∫mero (o 'fin' para terminar): ")
    
    if entrada.lower() == 'fin':
        break
    
    try:
        numero = float(entrada)
        numeros.append(numero)
        contador += 1
        print(f"N√∫mero {contador}: {numero}")
    except ValueError:
        print("Por favor, ingresa solo n√∫meros")

if len(numeros) > 0:
    print("\n--------------------------------------------------------------------")
    print("RESULTADOS:")
    print("--------------------------------------------------------------------\n")

    total = sum(numeros)
    promedio = total / len(numeros)
    
    maximo = max(numeros)
    minimo = min(numeros)
    
    print("- N√∫meros ingresados:", numeros)
    print(f"- Cantidad de n√∫meros: {len(numeros)}")
    print(f"- Suma total: {total}")
    print(f"- Promedio: {promedio:.2f}")
    print(f"- N√∫mero m√°s grande: {maximo}")
    print(f"- N√∫mero m√°s peque√±o: {minimo}")
    
else:
    print("\nNo ingresaste ning√∫n n√∫mero para analizar")

print("\n--------------------------------------------------------------------")
print("Fin de la calculadora")
```

##### **üîπ Ejercicio 5:** Contador de vocales

```python
print("     -> CONTADOR DE VOCALES")
print("--------------------------------------------------------------------")
print("Te dir√© cu√°ntas vocales tiene cualquier palabra u oraci√≥n")

texto = input("\nEscribe una palabra u oraci√≥n: ")
texto_minusculas = texto.lower()

vocales = 'aeiou√°√©√≠√≥√∫'

total_vocales = 0
contador_a = 0
contador_e = 0
contador_i = 0
contador_o = 0
contador_u = 0

for letra in texto_minusculas:
    if letra in vocales:
        total_vocales += 1
        
        if letra in 'a√°':
            contador_a += 1
        elif letra in 'e√©':
            contador_e += 1
        elif letra in 'i√≠':
            contador_i += 1
        elif letra in 'o√≥':
            contador_o += 1
        elif letra in 'u√∫':
            contador_u += 1

print("--------------------------------------------------------------------")
print("RESULTADOS:")
print("--------------------------------------------------------------------")

if total_vocales > 0:
    print(f"Total de vocales: {total_vocales}")
    print("\nDesglose por vocal:")
    print(f"   A: {contador_a}")
    print(f"   E: {contador_e}")
    print(f"   I: {contador_i}")
    print(f"   O: {contador_o}")
    print(f"   U: {contador_u}")

else:
    print("No se encontraron vocales en el texto")

print(f"\nTexto original: \"{texto}\"")
print("\n--------------------------------------------------------------------")
print("Fin del contador")
```

##### **ðŸ”¹ Ejercicio 1:** Conversor de temperaturas

```python
print("     -> CONVERSOR DE TEMPERATURA")
print("--------------------------------------------------------------------")

while True:
    print("\n1. Celsius a Fahrenheit")
    print("2. Fahrenheit a Celsius") 
    print("0. Salir")
    print("--------------------------------------------------------------------")
    
    opcion = input("Elige una opciÃ³n (0, 1 o 2): ")
    
    if opcion == "0":
        print("Fin del conversor")
        break
        
    elif opcion == "1":
        celsius = float(input("Ingresa los grados Celsius: "))
        fahrenheit = (celsius * 9/5) + 32
        print(f"{celsius}Â°C = {fahrenheit:.2f}Â°F")
        
    elif opcion == "2":
        fahrenheit = float(input("Ingresa los grados Fahrenheit: "))
        celsius = (fahrenheit - 32) * 5/9
        print(f"{fahrenheit}Â°F = {celsius:.2f}Â°C")
        
    else:
        print("OpciÃ³n incorrecta. Solo 0, 1 o 2.")
```

##### **ðŸ”¹ Ejercicio 2:** Contador de palabras en textos

```python
print("     -> CONTADOR DE PALABRAS")
print("--------------------------------------------------------------------")

while True:
    texto = input("Escribe un texto (o 'fin' para terminar): ")
    
    if texto.lower() == "fin":
        print("Fin del contador")
        break
    
    if texto.strip() == "":
        print("No escribiste ningÃºn texto. Intenta otra vez.")
        continue
    
    print(f"\nTu frase: \"{texto}\"")
    
    palabras = texto.split()
    palabras_validas = []
    
    for palabra in palabras:

        tiene_letras = False
        for caracter in palabra:
            if caracter.isalpha():  
                tiene_letras = True
                break
        
        if tiene_letras:
            palabras_validas.append(palabra)
    
    cantidad_palabras = len(palabras_validas)
    
    letras = 0
    for caracter in texto:
        if caracter.isalpha(): 
            letras += 1
    
    print(f"- Tu frase tiene {cantidad_palabras} palabras")
    print(f"- Tu frase tiene {letras} letras")
    print("--------------------------------------------------------------------")
```
